```python
import matplotlib.pyplot as plt
%matplotlib inline
import torch
import numpy as np
import numpy as np
from numpy.random import random
from mpl_toolkits.mplot3d import Axes3D

torch.__version__
```

```python
x = random((30,2))
weights = np.array([[2, -3]], dtype = np.float16)
b = np.array([1], dtype = np.float16)
```

```python
x.shape
```

```python
y = np.dot(x, [2., -3.]) + 1
y = y.reshape(-1, 1)
y
```

```python
# define learnable weights.
# random intialization for W and b

W = random(2).reshape(1, -1)
b_init = random(1).reshape(-1, 1)

print(f"intialized weights: {W}, {b_init}")
```

```python
def forward(W, b, X):
  return X @ W.T +b
y_hat = forward(W, b, x)
y_hat
```

```python
def loss(y_hat, y):
  return np.sum((y_hat - y)**2)
loss(y_hat, y)
```

```python
# dl/dW, dl/db_init
def grad(y_hat, y, X):
  dW = (y_hat - y).T @ X
  db = np.sum((y_hat - y))
  return 2*dW, 2*db
```

```python
grad(y_hat, y, x)
```

```python
lr = 0.001
for epoch in range(1000):
  y_hat = forward(W, b_init, x)
  l = loss(y_hat, y)
  dW, db = grad(y_hat, y, x)
  # gradient descent step
  W = W - lr * dW
  b_init = b_init - lr * db
  print("progress:", "epoch:", epoch, "loss", l)
# After training
print("estimation of the parameters:", W, b_init)
```
